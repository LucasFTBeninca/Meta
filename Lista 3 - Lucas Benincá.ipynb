```python
#Questão 07
soma = 0
contador = 0
while contador < 10:
    numero = int(input(f'Digite o {contador +1} º inteiro positivo: '))
    if numero > 0:
        soma += numero
        contador += 1
    else: 
        print('Número ignorado. Por favor, insira um inteiro positivo.')
media = soma / 10
print()
print(f'A média dos 10 inteiros positivos é: {media}')
```

```python
#Questão 08
menor = float('inf')
maior = float('-inf')
for i in range(1,11):
    numero = float(input(f'Digite o {i}º número: '))
    if numero < menor:
        menor = numero
    if numero > maior:
        maior = numero
print()
print(f'O menor valor lido é: {menor}')
print(f'O maior valor lido é: {maior}')
```
