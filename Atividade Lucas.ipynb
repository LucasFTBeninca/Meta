```python
# Importando o módulo de plotagem
import matplotlib.pyplot as plt

# Função para ler o arquivo .tsplib
def read_tsplib(filename):
    # Abre e lê o arquivo
    with open(filename, 'r') as file:
        lines = file.readlines()
        
    # Encontra o índice onde começam as coordenadas das cidades e onde termina o arquivo
    node_coord_start = lines.index('NODE_COORD_SECTION\n') + 1
    node_coord_end = lines.index('EOF\n')
    
    # Dicionário para armazenar as coordenadas das cidades
    cities = {}
    
    # Loop para extrair as coordenadas das cidades e armazenar no dicionário
    for line in lines[node_coord_start:node_coord_end]:
        index, x, y = map(float, line.strip().split())
        cities[int(index)] = (x, y)
    
    # Retorna o dicionário com as cidades e suas coordenadas
    return cities

# Função para plotar as cidades do arquivo .tsplib
def plot_tsplib(cities):
    # Cria um gráfico
    fig, ax = plt.subplots()
    
    # Plota as cidades como pontos no gráfico
    ax.scatter(*zip(*cities.values()), marker='o', color='red')
    
    # Adiciona anotações para cada cidade no gráfico
    for city, coords in cities.items():
        ax.annotate(city, coords)
    
    # Mostra o gráfico
    plt.show()

# Ponto de entrada do programa
if __name__ == '__main__':
    # Caminho do arquivo .tsplib (deve ser substituído pelo caminho correto)
    filename = 'xqf1.tsp'
    
    # Chama a função para ler o arquivo e armazena o resultado no dicionário 'cities'
    cities = read_tsplib(filename)
    
    # Chama a função para plotar, ao final do código, dá destaque à função de plotagem
    plot_tsplib(cities)
```

```python
import matplotlib.pyplot as plt
import random

# Função para ler os dados do arquivo TSPLIB e armazená-los em um dicionário.
def read_tsplib(filename):
    """Lê o arquivo .tsplib e retorna um dicionário com as cidades e suas coordenadas."""
    # Abre o arquivo para leitura.
    with open(filename, 'r') as file:
        lines = file.readlines()
        
    # Identifica a seção do arquivo que contém as coordenadas das cidades.
    node_coord_start = lines.index('NODE_COORD_SECTION\n') + 1
    node_coord_end = lines.index('EOF\n')
    
    cities = {}
    # Itera sobre as linhas da seção e armazena as coordenadas no dicionário.
    for line in lines[node_coord_start:node_coord_end]:
        index, x, y = map(float, line.strip().split())
        cities[int(index)] = (index, x, y)
    
    return cities

# Função para exibir um gráfico com as cidades.
def plot_tsplib(cities):
    """Plota as cidades do arquivo .tsplib."""
    # Inicializa a figura e os eixos.
    fig, ax = plt.subplots()
    # Plota as cidades como pontos vermelhos.
    ax.scatter([coords[1] for coords in cities.values()], [coords[2] for coords in cities.values()], marker='o', color='red')
    # Adiciona anotações (índices das cidades) ao gráfico.
    for city, coords in cities.items():
        ax.annotate(city, (coords[1], coords[2]))
    plt.show()

# Função para inicializar a população com rotas aleatórias.
def initialize_population(cities, population_size):
    """Inicializa uma população de rotas para o problema do Caixeiro Viajante."""
    population = []
    # Seleciona a primeira cidade como ponto de partida e chegada para todas as rotas.
    origin_city = list(cities.keys())[0]
    for _ in range(population_size):
        # Cria uma rota aleatória a partir das outras cidades.
        route = random.sample(list(cities.keys())[1:], len(cities)-1)
        # Adiciona a cidade de origem ao início e ao fim da rota.
        route.insert(0, origin_city)
        route.append(origin_city)
        population.append(route)
    
    return population

# Função para calcular a distância entre duas cidades.
def calcular_distancia_entre_cidades(cidade1, cidade2):
    """Calcula a distância euclidiana entre duas cidades."""
    x1, y1 = cidade1[1], cidade1[2]
    x2, y2 = cidade2[1], cidade2[2]
    distancia = ((x2 - x1) ** 2 + (y2 - y1) ** 2) ** 0.5
    return distancia

# Função para calcular a distância total de uma rota.
def calcular_distancia_total(rota, cities):
    """Calcula a distância total de uma rota específica."""
    distancia_total = 0.0
    for i in range(len(rota) - 1):
        cidade_atual = cities[rota[i]]
        proxima_cidade = cities[rota[i + 1]]
        distancia = calcular_distancia_entre_cidades(cidade_atual, proxima_cidade)
        distancia_total += distancia
    # Adiciona a distância de retorno ao ponto de partida.
    distancia_total += calcular_distancia_entre_cidades(cities[rota[-1]], cities[rota[0]])
    return distancia_total

# Função para calcular a aptidão (ou custo) de uma rota.
def calcular_aptidao(cromossomo, cities):
    """Calcula a aptidão de um cromossomo (rota) usando a distância total da rota."""
    return calcular_distancia_total(cromossomo, cities)

# Função principal
if __name__ == '__main__':
    # Nome do arquivo a ser lido.
    filename = 'xqf1.tsp'
    # Carrega as cidades do arquivo.
    cities = read_tsplib(filename)
    # Exibe as cidades em um gráfico.
    plot_tsplib(cities)
    
    # Define o tamanho da população.
    population_size = 50
    # Inicializa a população.
    initial_population = initialize_population(cities, population_size)
    # Para cada rota na população inicial, calcula e imprime sua aptidão.
    for rota in initial_population:
        aptidao = calcular_aptidao(rota, cities)
        print(f"Aptidão da rota: {aptidao:.2f}")
```

```python
import matplotlib.pyplot as plt

# Função para ler os dados do arquivo TSPLIB e armazená-los em um dicionário
def read_tsplib(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
        
    # Localiza as seções de coordenação e EOF no arquivo
    node_coord_start = lines.index('NODE_COORD_SECTION\n') + 1
    node_coord_end = lines.index('EOF\n')
    
    cities = {}
    # Preenche o dicionário com as informações do arquivo
    for line in lines[node_coord_start:node_coord_end]:
        index, x, y = map(float, line.strip().split())
        cities[int(index)] = (index, x, y)
    
    return cities

# Função para plotar as cidades a partir das coordenadas no gráfico
def plot_tsplib(cities):
    fig, ax = plt.subplots()
    x_vals = [coords[1] for coords in cities.values()]
    y_vals = [coords[2] for coords in cities.values()]
    ax.scatter(x_vals, y_vals, marker='o', color='red')
    
    # Anota o índice da cidade ao lado do ponto no gráfico
    for city, coords in cities.items():
        ax.annotate(city, (coords[1], coords[2]))
    
    plt.show()

# Calcula a distância entre duas cidades usando a fórmula de distância Euclidiana
def calcular_distancia_entre_cidades(cidade1, cidade2):
    x1, y1 = cidade1[1], cidade1[2]
    x2, y2 = cidade2[1], cidade2[2]
    distancia = ((x2 - x1) ** 2 + (y2 - y1) ** 2) ** 0.5
    return distancia

# Calcula a distância total de uma rota específica
def calcular_distancia_total(rota, cities):
    distancia_total = 0.0
    for i in range(len(rota) - 1):
        cidade_atual = cities[rota[i]]
        proxima_cidade = cities[rota[i + 1]]
        distancia = calcular_distancia_entre_cidades(cidade_atual, proxima_cidade)
        distancia_total += distancia
    # Adiciona a distância entre a última cidade e a cidade inicial para completar o ciclo
    distancia_total += calcular_distancia_entre_cidades(cities[rota[-1]], cities[rota[0]])
    return distancia_total

# Implementa o algoritmo do Vizinho Mais Próximo para encontrar uma solução aproximada
def vizinho_mais_proximo(cities):
    cidade_atual = list(cities.keys())[0]
    rota = [cidade_atual]
    cidades_nao_visitadas = set(cities.keys())
    cidades_nao_visitadas.remove(cidade_atual)

    # Continua selecionando a cidade mais próxima da cidade atual até que todas as cidades sejam visitadas
    while cidades_nao_visitadas:
        proxima_cidade = min(cidades_nao_visitadas, key=lambda cidade: calcular_distancia_entre_cidades(cities[cidade_atual], cities[cidade]))
        rota.append(proxima_cidade)
        cidades_nao_visitadas.remove(proxima_cidade)
        cidade_atual = proxima_cidade

    rota.append(rota[0])  # Retorna à cidade inicial para completar o ciclo
    return rota

# Função para plotar a rota no gráfico
def plot_rota(cities, rota):
    rota_coords = [cities[city] for city in rota]
    x_vals = [coord[1] for coord in rota_coords]
    y_vals = [coord[2] for coord in rota_coords]

    plt.figure(figsize=(10, 5))
    plt.plot(x_vals, y_vals, '-o')
    plt.show()

# Principal
if __name__ == '__main__':
    filename = 'xqf1.tsp'
    cities = read_tsplib(filename)
    plot_tsplib(cities)  # Plota as cidades
    
    # Encontra a melhor rota usando o Vizinho Mais Próximo
    melhor_rota = vizinho_mais_proximo(cities)
    distancia = calcular_distancia_total(melhor_rota, cities)
    
    # Plota a rota e imprime os resultados
    plot_rota(cities, melhor_rota)
    print(f"Distância total da rota encontrada: {distancia:.2f}")
    print("Rota:", melhor_rota)
```
